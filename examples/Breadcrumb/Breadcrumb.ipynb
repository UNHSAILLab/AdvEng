## Breadcrumb
This simple example outlines how to describe a simple and bare example of how to the breadcrumb ae object and attach objectives to `lureObjects`.

In this example, you will see how to use:
- `engage:Breadcrumb`
- `engage:Stage`
- `Objective:Lure`



breadcrumb.png




Additionally there is the `engage:BreadcrumbTrail` object in case where the breadcrumbs themselves have some partial ordering. Alternatively, one can describe `engage:Breadcrumb` with multiple `uco-core:UcoObject(s)` if all components as a whole is the breadcrumb concept.

```python
{
    "@context": {
        "co": "http://purl.org/co/",
        "kb": "http://example.org/kb/",
        "rdfs": "http://www.w3.org/2000/01/rdf-schema#",
        "uco-observable": "https://ontology.unifiedcyberontology.org/uco/observable/",
        "uco-core": "https://ontology.unifiedcyberontology.org/uco/core/",
        "uco-types": "https://ontology.unifiedcyberontology.org/uco/types/",
        "uco-identity": "https://ontology.unifiedcyberontology.org/uco/identity/", 
        "uco-action": "https://ontology.unifiedcyberontology.org/uco/action/", 
		"engage": "https://ontology.adversaryengagement.org/ae/engage/",
		"objective": "https://ontology.adversaryengagement.org/ae/objective/",
        "xsd": "http://www.w3.org/2001/XMLSchema#"
    },
    "@graph": [
				{
				"@id":"kb:bcdeployment1",
				"@type":"engage:Deploy",
				"uco-action:performer": [
					{
					"@type":"uco-identity:Person",
					"uco-core:hasFacet":[
						{
						"@id":"kb:LiamWilliam",
						"@type":"identity:SimpleNameFacet",
						"uco-identity:familyName":"Liam",
						"uco-identity:givenName":"William"
						}
						],
				"uco-action:object": [
					{
					"@id":"kb:breadcrumb1",
					"@type":"engage:Breadcrumb",
					"uco-core:description":"breadcrumb 1",
					"engage:hasObjective": [
						{
							"@id":"kb:lure-to-honeypot",
							"@type":"objective:Lure"
						}
						]
					},
					{
					"@id":"kb:breadcrumb2",
					"@type":"engage:Breadcrumb",
					"uco-core:description":"breadcrumb 2",
					"engage:hasObjective": [
						{
							"@id":"kb:lure-to-honeypot",
							"@type":"objective:Lure"
						}
						]
					},
					{
					"@id":"kb:breadcrumb3",
					"@type":"engage:Breadcrumb",
					"uco-core:description":"breadcrumb 3",
					"engage:hasObjective": [
						{
							"@id":"kb:lure-to-honeypot",
							"@type":"objective:Lure"
						}
						]
					}
					]
					}
				]
				},
				{
				"@type":"uco-core:Relationship",
				"uco-core:kindofRelationship":"has_stage",
				"uco-core:source":[
				{
				"@id":"kb:breadcrumb1",
				"@type":"engage:Breadcrumb"
				}
				],
				"uco-core:target":[
				{
					"@type":"engage:Stage",
					"engage:hasCharacteristics":[
					{
					"@id":"kb:dir1",
					"@type": "uco-observable:Directory",
					"uco-core:name": "C://marysu/projectBoat"
					}
					]
				}
				]
				},
				{
				"@type":"uco-core:Relationship",
				"uco-core:kindofRelationship":"has_stage",
				"uco-core:source":[
				{
				"@id":"kb:breadcrumb1",
				"@type":"engage:Breadcrumb"
				}
				],
				"uco-core:target":[
				{
					"@type":"engage:Stage",
					"engage:hasCharacteristics":[
					{
						"@id":"kb:file1",
						"@type":"uco-observable:File",
						"uco-core:name": "projectBoat.doc"
					}
					]
				}
				]
				},
				{
				"@type":"uco-core:Relationship",
				"uco-core:kindofRelationship":"has_stage",
				"uco-core:source":[
				{
				"@id":"kb:breadcrumb1",
				"@type":"engage:Breadcrumb"
				}
				],
				"uco-core:target":[
				{
					"@type":"engage:Stage",
					"engage:hasCharacteristics":[
					{
						"@id":"kb:file2",
						"@type":"uco-observable:PDFFile",
						"uco-core:name": "to-do-list.pdf"
					}
					]
				}
				]
				}
    ]
        
}
```
